```python
import matplotlib.pyplot as plt
import requests
import random
```

```python
data_url = "https://raw.githubusercontent.com/sixhobbits/ritza/master/data/us-cities.txt"
```

```python
r = requests.get(data_url)
```

```python
with open("us-cities.txt", "w") as f:
    f.write(r.text)
```

```python
lats = []
lons = []
colours = []
state_colours = {}
all_colours = ['blue', 'green', 'red', 'cyan', 'maroon', 'yellow', 'black',
               'aqua','lime','crimson','orange','pink','peru','coral']
with open("us-cities.txt") as f:
    for i, line in enumerate(f):
        state, lat, lon = line.split()
        lats.append(float(lat))
        lons.append(float(lon))
        
        if state not in state_colours:
            state_colours[state] = random.choice(all_colours)
        colours.append(state_colours[state])

plt.scatter(lons,lats,color=colours)
```
