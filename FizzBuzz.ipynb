```python
for i in range(1,101):
    if (i%3 == False) and not (i%5 == False):
        print('Fizz')
    
    if (i%5 == False) and not (i%3 == False):
        print('Buzz')
        
    elif (i%5 == False) and (i%3 == False):
        print('FizzBuzz')
        
    else:
        print(i)


    
    
```

```python
for i in range(1,101):
    output = ""
    if i%3 == 0:
        output += "Fizz"
    if i%5 == 0:
        output += "Buzz"
    if i%7 == 0:
        output += "Fuzz"
    if i%13 == 0:
        output += "Bizz"
    if output =="":
        output = i
    print(output)
```
